```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
plt.plot
df = pd.read_csv('subject_1432.csv', header=[0], skiprows=[1,2,3])

x = df['TIME']
y1 = df['VO2']
y2 = df['VE']
y1max = max(y1)
xmax = x[y1.argmax()]

fig, ax = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace=0)

ax[0].annotate(f'V\u0307O\u2082max={y1max:.2f}L/min', xy=(xmax,y1max), xytext=(xmax +.5, y1max +.5), 
                arrowprops=dict(facecolor='red', shrink=0.05))
ax[0].plot(x, y1, 'o', label='V\u0307O\u2082', c='b')
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].set(ylabel='L/min')
ax[0].legend()
ax[0].set_title('Exercise Respiration')

ax[1].plot(x, y2, 'o', label='VE', c='r')
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].set(ylabel='breaths/min')
ax[1].set(xlabel='Time(min)')
ax[1].legend()

fig.savefig('VO2-VE.png')
plt.show

x2 = df['VO2']
y3 = df['FECO2']
y4 = df['VCO2']

fig, ax = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace=0)

ax[0].plot(x2, y3, 'o', label='FECO\u2082', c='g')
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].set(ylabel='%')
ax[0].legend()
ax[0].set_title('CO\u2082 Levels during Exercise')
ax[0].axvline(x=1.95, linewidth=2)
ax[0].axvline(x=3.1, linewidth=2)

ax[1].plot(x2, y4, 'o', label='VCO\u2082', c='y')
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].set(ylabel='L/min')
ax[1].set(xlabel='VO\u2082(L/min)')
ax[1].legend()
ax[1].axvline(x=1.95, linewidth=2)
ax[1].axvline(x=3.1, linewidth=2)
fig.savefig('FECO2-VCO2.png')
plt.show
```
